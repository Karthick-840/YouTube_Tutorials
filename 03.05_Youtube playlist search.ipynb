```python
import os
from googleapiclient.discovery import build
import time
import datetime
import pandas as pd
import pafy
import pytube
from pytube import Playlist
from pafy import get_playlist
```

```python
f =  open('Links.txt','r')
content =  (f.read()).splitlines()
f.close()
```

```python
filtered = [k for k in content if 'www.youtube.com' in k]
only_videos = [k.split("&list=",1)[0] for k in filtered if 'watch?' in k]
only_playlists = [k for k in filtered if 'playlist?' in k]
```

```python
def extract(url,Type):
    if(Type=="video"):
        v = pafy.new(url)
        case = {'Title':v.title,'Duration':v.duration,'Ratings':v.rating,'Author':v.author,'Views':v.viewcount,'URL':url}
        case['Type'] = Type
    if(Type=="Playlist"):
        videos= Playlist(url)
        case =[]
        for video in videos:
            c = extract(video,"video")
            c['Type'] = Type
            case.append(c)
    return(case)
```

```python
Details =[]
for url in filtered:
    if('playlist?' in url):
        Details.extend(extract(url,'Playlist'))
    else:
        Details.append(extract(url,'video'))
```

```python
First_list = pd.DataFrame(Details)
First_list
```

```python
First_list.to_csv('Videos_playtime.csv')
```
